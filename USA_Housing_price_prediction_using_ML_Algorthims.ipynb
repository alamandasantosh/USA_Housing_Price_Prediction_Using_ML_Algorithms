```python
threshold = 0.5
def interpret_correlation(df, threshold):
  if abs(df) < threshold:
    return "No correlation"
  elif df > threshold:
    return "Positive correlation"
  else:
    return "Negative correlation"
corr_matrix = pd.DataFrame()
for col in df.columns:
  corr_matrix[col] = df[col].apply(lambda x: interpret_correlation(x, threshold))
corr_matrix
```
